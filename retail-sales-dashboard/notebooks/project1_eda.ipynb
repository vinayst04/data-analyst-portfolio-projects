# Retail Sales EDA
Simple checks used before building the Power BI dashboard.


```python
import pandas as pd

df = pd.read_csv('../data/processed/superstore_clean.csv')
df.head()
```

```python
print('Rows:', len(df))
print('Columns:', len(df.columns))
print(df.columns.tolist())
```

```python
total_sales = df['sales'].sum()
total_profit = df['profit'].sum()
profit_margin = total_profit / total_sales * 100
total_orders = df['order_id'].nunique()

print('Total Sales:', round(total_sales, 2))
print('Total Profit:', round(total_profit, 2))
print('Profit Margin %:', round(profit_margin, 2))
print('Total Orders:', total_orders)
```

```python
category_summary = df.groupby('category')[['sales', 'profit']].sum()
category_summary = category_summary.sort_values('sales', ascending=False)
category_summary
```

```python
sub_category_summary = df.groupby('sub_category')[['sales', 'profit']].sum()
sub_category_summary = sub_category_summary.sort_values('profit')
sub_category_summary.head(10)
```

```python
discount_summary = df.groupby('discount_band')[['sales', 'profit']].sum()
discount_summary = discount_summary.sort_values('profit')
discount_summary
```
