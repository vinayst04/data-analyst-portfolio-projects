# Airline Customer Sentiment EDA



```python
import pandas as pd

df = pd.read_csv('../data/processed/airline_sentiment_clean.csv')
df.head()
```

```python
df.shape
df['sentiment'].value_counts(normalize=True).mul(100).round(2)
```

```python
airline_negative_rate = df.groupby('airline')['is_negative'].mean()
airline_negative_rate = airline_negative_rate * 100
airline_negative_rate = airline_negative_rate.sort_values(ascending=False)
airline_negative_rate = airline_negative_rate.round(2)
airline_negative_rate
```
