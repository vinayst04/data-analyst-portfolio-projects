# Customer Churn EDA
Simple checks used before building the Power BI dashboard.


```python
import pandas as pd

df = pd.read_csv('../data/processed/telco_churn_clean.csv')
df.head()
```

```python
print('Rows:', len(df))
print('Columns:', len(df.columns))
print(df.columns.tolist())
```

```python
total_customers = len(df)
churned_customers = df['churn_flag'].sum()
churn_rate = churned_customers / total_customers * 100
monthly_risk = df['monthly_revenue_at_risk'].sum()

print('Total customers:', total_customers)
print('Churned customers:', churned_customers)
print('Churn rate %:', round(churn_rate, 2))
print('Monthly revenue at risk:', round(monthly_risk, 2))
```

```python
contract_churn = df.groupby('contract')['churn_flag'].mean()
contract_churn = contract_churn * 100
contract_churn = contract_churn.sort_values(ascending=False)
contract_churn
```

```python
tenure_churn = df.groupby('tenure_group')['churn_flag'].mean()
tenure_churn = tenure_churn * 100
tenure_churn
```

```python
payment_churn = df.groupby('payment_method')['churn_flag'].mean()
payment_churn = payment_churn * 100
payment_churn = payment_churn.sort_values(ascending=False)
payment_churn
```
